```python
import random 
while(True):
     t=random.randint(10,99)
     h=random.randint(10,99)
     if(t>30 and h<40):
       print("High temperature and values of Temperature & Humidity is : ",t,h,"Alarm is on")
     elif(t<30 and h>40):
       print("Low temperature and values of Temperature & Humidity : ",t,h,"Alarm is off")
```
